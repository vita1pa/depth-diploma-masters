```python
import numpy as np
import matplotlib.pyplot as plt

def F_MC(theta, phi, lambda1, lambda2, N=100000):
    """
    Monte Carlo estimate of F_Y(t) where 
      Y = cos(phi)*X1 + sin(phi)*X2,   X1 ~ Exp(lambda1), X2 ~ Exp(lambda2),
    and t = theta dot (cos(phi), sin(phi)).
    
    Since X1 and X2 are supported on [0, ∞), the law of Y depends on the signs
    of a=cos(phi) and b=sin(phi). This function estimates
      F_Y(t) = P( a*X1 + b*X2 <= t )
    for any t in R.
    """
    a = np.cos(phi)
    b = np.sin(phi)
    t = theta[0]*a + theta[1]*b  # projection of theta in direction phi

    # Generate N samples from the independent exponentials
    # Note: np.random.exponential uses the "scale" parameter = 1/lambda.
    X1 = np.random.exponential(scale=1/lambda1, size=N)
    X2 = np.random.exponential(scale=1/lambda2, size=N)
    
    # Compute Y = a*X1 + b*X2. Note: depending on phi, a and/or b can be negative.
    Y = a * X1 + b * X2
    
    # Estimate the CDF at t
    F_est = np.mean(Y <= t)
    return F_est

def one_minus_F_MC(theta, phi, lambda1, lambda2, N=100000):
    """
    Computes 1 - F_Y(t) via Monte Carlo, with t = theta dot (cos(phi), sin(phi)).
    """
    return 1 - F_MC(theta, phi, lambda1, lambda2, N)

def plot_one_minus_F_MC(theta, lambda1, lambda2, phi_min=0, phi_max=2*np.pi, num_points=200, N=100000):
    """
    Iterates through phi values from phi_min to phi_max (default full circle)
    and plots 1 - F_Y(t) as a function of phi.
    """
    phi_grid = np.linspace(phi_min, phi_max, num_points)
    values = np.zeros_like(phi_grid)
    
    # For reproducibility, you might fix a random seed here.
    np.random.seed(0)
    
    # Pre-generate samples from the joint law of X.
    X1 = np.random.exponential(scale=1/lambda1, size=N)
    X2 = np.random.exponential(scale=1/lambda2, size=N)
    
    for i, phi in enumerate(phi_grid):
        a = np.cos(phi)
        b = np.sin(phi)
        t = theta[0]*a + theta[1]*b  # projected threshold
        Y = a * X1 + b * X2
        F_est = np.mean(Y <= t)
        values[i] = 1 - F_est
        
    plt.figure(figsize=(8, 5))
    plt.plot(phi_grid, values, label=r'$1-F(\theta)$')
    plt.xlabel(r'$\varphi$ (radians)')
    plt.ylabel(r'$1-F(\theta)$')
    plt.title(f'Variation of $1-F(\\theta)$ with direction for $\\theta={theta}$\n'
              f'with $\\lambda_1={lambda1}$, $\\lambda_2={lambda2}$')
    plt.legend()
    plt.grid(True)
    plt.show()

def find_optimal_phi_MC(theta, lambda1, lambda2, phi_min=0, phi_max=2*np.pi, num_points=200, N=100000):
    """
    Finds the angle phi (in [phi_min, phi_max]) at which 1 - F_Y(t) is minimized,
    using Monte Carlo estimates.
    Returns the optimal phi and the corresponding minimum value.
    """
    phi_grid = np.linspace(phi_min, phi_max, num_points)
    values = np.zeros_like(phi_grid)
    
    # Pre-generate samples.
    np.random.seed(0)
    X1 = np.random.exponential(scale=1/lambda1, size=N)
    X2 = np.random.exponential(scale=1/lambda2, size=N)
    
    for i, phi in enumerate(phi_grid):
        a = np.cos(phi)
        b = np.sin(phi)
        t = theta[0]*a + theta[1]*b
        Y = a * X1 + b * X2
        F_est = np.mean(Y <= t)
        values[i] = 1 - F_est
    
    idx_min = np.argmin(values)
    return phi_grid[idx_min], values[idx_min]

# Example usage:
if __name__ == '__main__':
    # Fix a point theta in R^2 (choose any point you like)
    theta = np.array([0.79, 0.79])
    
    # Define arbitrary rate parameters
    lambda1 = 1
    lambda2 = 1
    
    # Plot the variation of 1 - F(theta) as the direction rotates through [0, 2pi]
    plot_one_minus_F_MC(theta, lambda1, lambda2, phi_min=0, phi_max=2*np.pi, num_points=200, N=100000)
    
    # Find the optimal direction phi (that minimizes 1 - F(theta))
    opt_phi, opt_val = find_optimal_phi_MC(theta, lambda1, lambda2, phi_min=0, phi_max=2*np.pi, num_points=200, N=100000)
    print(f"Optimal phi (radians): {opt_phi:.4f}, with 1-F(theta) = {opt_val:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

def one_minus_F_MC(theta, phi, lambda1, lambda2, N=25000):
    """
    Monte Carlo estimate of 1 - F_Y(t), where:
      - Y = cos(phi)*X1 + sin(phi)*X2
      - t = theta dot (cos(phi), sin(phi))
      - X1 ~ Exp(lambda1), X2 ~ Exp(lambda2)
    """
    a = np.cos(phi)
    b = np.sin(phi)
    t = theta[0] * a + theta[1] * b  # projection of theta in direction (a, b)

    # Sample from independent exponential distributions
    X1 = np.random.exponential(scale=1 / lambda1, size=N)
    X2 = np.random.exponential(scale=1 / lambda2, size=N)

    # Compute the linear combination Y
    Y = a * X1 + b * X2

    # Estimate F_Y(t) = P(Y <= t)
    F_est = np.mean(Y <= t)

    # Return 1 - F_Y(t)
    return 1 - F_est

def find_optimal_phi_dichotomy(theta, lambda1, lambda2,
                               phi_min=0, phi_max=2*np.pi,
                               tol=1e-4, max_iter=50, N=50000):
    """
    Finds the optimal phi in [phi_min, phi_max] that minimizes:
      f(phi) = one_minus_F_MC(theta, phi, lambda1, lambda2, N),
    using a dichotomy (bisection) search.
    """
    a_phi = phi_min
    b_phi = phi_max
    delta = tol / 2.0

    for _ in range(max_iter):
        mid = (a_phi + b_phi) / 2.0
        phi1 = mid - delta
        phi2 = mid + delta

        f1 = one_minus_F_MC(theta, phi1, lambda1, lambda2, N)
        f2 = one_minus_F_MC(theta, phi2, lambda1, lambda2, N)

        if f1 < f2:
            b_phi = phi2
        else:
            a_phi = phi1

        if (b_phi - a_phi) < tol:
            break

    phi_opt = (a_phi + b_phi) / 2.0
    f_opt = one_minus_F_MC(theta, phi_opt, lambda1, lambda2, N)
    return phi_opt, f_opt

def compute_depth(theta, lambda1, lambda2,
                  coarse_points=100, N_coarse=5000,
                  tol=1e-4, N_dicho=10000):
    """
    Estimates the Tukey depth for a fixed point theta in R^2, where:
      D(theta) = min_{phi in [0, 2pi]} [1 - F_Y(theta dot u(phi))].
    1) Performs a coarse grid search over phi to find an approximate minimizer.
    2) Refines it using the dichotomy method in a small neighborhood.
    Returns (depth_value, phi_opt).
    """
    phi_grid = np.linspace(0, 2*np.pi, coarse_points)
    f_vals = np.zeros_like(phi_grid)

    # Fix the seed to reduce variance in the coarse search
    np.random.seed(0)
    for i, phi in enumerate(phi_grid):
        f_vals[i] = one_minus_F_MC(theta, phi, lambda1, lambda2, N_coarse)

    idx_min = np.argmin(f_vals)
    phi_coarse = phi_grid[idx_min]
    f_coarse = f_vals[idx_min]

    # Define a small interval around the coarse optimum
    delta_interval = 0.2
    a_phi = phi_coarse - delta_interval
    b_phi = phi_coarse + delta_interval

    # Clamp to [0, 2*pi]
    if a_phi < 0:
        a_phi = 0
    if b_phi > 2*np.pi:
        b_phi = 2*np.pi

    # Refine using dichotomy search
    phi_opt, f_opt = find_optimal_phi_dichotomy(
        theta, lambda1, lambda2,
        phi_min=a_phi, phi_max=b_phi,
        tol=tol, N=N_dicho
    )

    # Choose the best among the coarse and refined estimates
    if f_opt > f_coarse:
        return f_coarse, phi_coarse
    else:
        return f_opt, phi_opt

def contour_depth_plot(lambda1, lambda2, theta1_range, theta2_range,
                       grid_points=30, coarse_points=100,
                       N_coarse=5000, tol=1e-4, N_dicho=10000):
    """
    Computes the Tukey depth for points (theta1, theta2) over a grid
    in the rectangle [theta1_range] x [theta2_range].
    Then creates a contour plot of the depth values.
    """
    theta1_vals = np.linspace(theta1_range[0], theta1_range[1], grid_points)
    theta2_vals = np.linspace(theta2_range[0], theta2_range[1], grid_points)

    Depth = np.zeros((grid_points, grid_points))
    Phi_opt = np.zeros((grid_points, grid_points))  # store the optimal phi

    # Outer loop with progress bar over theta1 values
    for i, th1 in enumerate(tqdm(theta1_vals, desc="Processing theta1 values")):
        for j, th2 in enumerate(theta2_vals):
            theta = np.array([th1, th2])
            d_val, phi_val = compute_depth(
                theta, lambda1, lambda2,
                coarse_points=coarse_points,
                N_coarse=N_coarse, tol=tol, N_dicho=N_dicho
            )
            Depth[j, i] = d_val   # j for y-axis, i for x-axis
            Phi_opt[j, i] = phi_val

    # Create the contour plot
    T1, T2 = np.meshgrid(theta1_vals, theta2_vals)
    plt.figure(figsize=(8, 6))
    cp = plt.contourf(T1, T2, Depth, levels=20, cmap='viridis')
    plt.colorbar(cp, label='Tukey Depth (min 1 - F)')
    plt.xlabel(r'$\theta_1$')
    plt.ylabel(r'$\theta_2$')
    plt.title(f'Contour of Tukey Depth (Exp($\\lambda_1={lambda1}$, $\\lambda_2={lambda2}$))')

    # Optionally, save the figure
    # plt.savefig("tukey_depth.png", dpi=300)
    plt.show()

    return T1, T2, Depth, Phi_opt

# Example usage
if __name__ == '__main__':
    # Define rate parameters for the exponentials
    lambda1 = 0.25
    lambda2 = 0.75

    # Define a grid region in theta-space
    theta1_range = [0, 3]
    theta2_range = [0, 3]

    # Generate and display the contour plot
    T1, T2, Depth, Phi_opt = contour_depth_plot(
        lambda1, lambda2,
        theta1_range, theta2_range,
        grid_points=100  # instead of 30
    )

    # Print the global minimum depth and its location
    min_idx = np.unravel_index(np.argmin(Depth), Depth.shape)
    theta_min = (T1[min_idx], T2[min_idx])
    print(f"Global minimum depth ~ {Depth[min_idx]:.4f} at theta = {theta_min}")
```
