```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the point (a,b)
a, b = 2, 3

# Set up a range of x-values for plotting
x_vals = np.linspace(0, 8, 400)

# Create a figure with three subplots for the three cases
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
# fig.suptitle("Visualization of Regions Defined by Lines Through (a,b)", fontsize=16)

# --------- Case 1: Region I (Triangle area) ---------
# Line 1: choose m1 = -1 so that the line has positive intercepts.
m1 = -1
# Equation: y = m1*(x-a)+b
y_line1 = m1 * (x_vals - a) + b  # = -x + 5
# The intercepts are at x=5 (x-axis) and y=5 (y-axis)
x_intercept1 = 5
# For the triangular region, restrict x to [0, x_intercept1]
x_tri = np.linspace(0, x_intercept1, 200)
y_tri = m1 * (x_tri - a) + b  # = -x_tri + 5

ax = axs[0]
ax.plot(x_vals, y_line1, "b-", label="Line 1: y = -x+5")
ax.fill_between(x_tri, 0, y_tri, color="blue", alpha=0.3)
ax.plot(a, b, "ro", label="Point (a,b)")
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
# ax.set_title("Region I (Triangle)")
# ax.legend()
# Annotate the region
ax.text(1, 3, "I", fontsize=20, color="blue", weight="bold")

# --------- Case 2: Region II (x ≥ 0, 0 ≤ y ≤ m(x-a)+b) ---------
# Line 2: choose m2 = 1 so that the x-intercept is negative and y-intercept positive.
m2 = 1
# Equation: y = m2*(x-a)+b = x-2+3 = x+1
y_line2 = m2 * (x_vals - a) + b  # = x_vals + 1

ax = axs[1]
ax.plot(x_vals, y_line2, "g-", label="Line 2: y = x+1")
# For x >= 0, fill the region from y=0 up to the line.
ax.fill_between(x_vals, 0, y_line2, where=(x_vals >= 0), color="green", alpha=0.3)
ax.plot(a, b, "ro", label="Point (a,b)")
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
# ax.set_title("Region II (Below Line)")
# ax.legend()
ax.text(1, 1.5, "II", fontsize=20, color="green", weight="bold")

# --------- Case 3: Region III (x ≥ x₀, 0 ≤ y ≤ m(x-a)+b) ---------
# Line 3: choose m3 = 2 so that the x-intercept is positive and y-intercept negative.
m3 = 2
# Equation: y = m3*(x-a)+b = 2*(x-2)+3 = 2x - 1
y_line3 = m3 * (x_vals - a) + b  # = 2*x_vals - 1
# x-intercept is when y=0, i.e., 2x - 1 = 0 --> x = 0.5
x_intercept3 = 0.5

ax = axs[2]
ax.plot(x_vals, y_line3, "r-", label="Line 3: y = 2x-1")
# For x >= x_intercept3, fill the region from y=0 up to the line.
mask = x_vals >= x_intercept3
ax.fill_between(x_vals[mask], 0, y_line3[mask], color="red", alpha=0.3)
ax.plot(a, b, "ko", label="Point (a,b)")
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
# ax.set_title("Region III (Under Line, x ≥ x₀)")
# ax.legend()
ax.text(1, 0.5, "III", fontsize=20, color="red", weight="bold")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
